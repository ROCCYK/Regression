2- Drop the following columns: Mothers job (Mjob), Fathers job (Fjob), Reason at School
(reason), and Students Guardian (guardian) (5 Marks).

```python
#Question 2:
df = df.drop(columns=['Mjob','Fjob','reason','guardian'])
df
```

3- Check if there are null values and fill them with the average value of that attribute if it is
numerical, and with the most common occurrence of that value if it is categorical (5
Marks).

```python
#Question 3:
df.isnull().sum()
```

```python
#Question 3:
percentage = 0.01
mask = np.random.choice([True, False], size=df.shape, p=[percentage,1-percentage])
dfmod1= df.mask(mask)

dfmod1.isnull().sum()
```

```python
#Question 3:
dfmod1[['age','Medu','Fedu','traveltime','studytime','failures','famrel','freetime','goout','Dalc','Walc','health','absences','G1','G2','G3']] = dfmod1[['age','Medu','Fedu','traveltime','studytime','failures','famrel','freetime','goout','Dalc','Walc','health','absences','G1','G2','G3']].apply(lambda x: x.fillna(x.mean()))
dfmod1[['school','sex','address','famsize','Pstatus','schoolsup','famsup','paid','activities','nursery','higher','internet','romantic']] = dfmod1[['school','sex','address','famsize','Pstatus','schoolsup','famsup','paid','activities','nursery','higher','internet','romantic']].apply(lambda x: x.fillna(x.mode()[0]))
```

```python
#Question 3:
dfmod1.isnull().sum()
```

4- Encode the categorical columns using get_dummies encoder (10 Marks).

```python
#Question 4:
dfmod2 = pd.get_dummies(dfmod1)
dfmod2
```

5- Split the data into 80% training set and 20% testing set (10 Marks).

```python
#Question 5:
from sklearn.model_selection import train_test_split

X = dfmod2.iloc[:,dfmod2.columns != 'G3']
y = dfmod2['G3']
SEED = 4

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.20,random_state=SEED)
```

6- Train a linear regression model to predict G3 (10 Marks).

```python
#Question 6:
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg.fit(X_train,y_train)

y_pred = reg.predict(X_test)
```

7- Test on the test dataset, and report on the Mean Absolute Error (MAE) and Mean
Squared Error (MSE). Compare predicted sample results (e.g., 5 samples) with the test
labels. (10 Marks).

```python
#Question 7:
from sklearn.metrics import mean_absolute_error, mean_squared_error

print(f'MAE: {mean_absolute_error(y_test,y_pred)}')
print(f'MSE: {mean_squared_error(y_test,y_pred)}')

result = pd.DataFrame({'Prediction':y_pred,'Actual':y_test})
print(result)
```

8- Train a Ridge regression model and a Lasso (alpha=0.5) regression model (10 Marks).

```python
#Question 8:
from sklearn.linear_model import Ridge, Lasso

ridge = Ridge(alpha=0.5)
ridge.fit(X_train,y_train)

y_pred_ridge = ridge.predict(X_test)
```

```python
#Question 8:
lasso = Lasso(alpha=0.5)
lasso.fit(X_train,y_train)

y_pred_lasso = lasso.predict(X_test)
```

9- Test the models trained in step 8 on the test dataset and compare the MAE and MSE
results with the ones you got in step 7 (10 Marks).

```python
#Question 9:
print(f'Linear Regression MAE: {mean_absolute_error(y_test,y_pred)}')
print(f'Linear Regression MSE: {mean_squared_error(y_test,y_pred)}')

print(f'Ridge Regression MAE: {mean_absolute_error(y_test,y_pred_ridge)}')
print(f'Ridge Regression MSE: {mean_squared_error(y_test,y_pred_ridge)}')

print(f'Lasso Regression MAE: {mean_absolute_error(y_test,y_pred_lasso)}')
print(f'Lasso Regression MSE: {mean_squared_error(y_test,y_pred_lasso)}')

compare = pd.DataFrame({'Linear Regression Prediction':y_pred,'Ridge Regression Prediction':y_pred_ridge,'Lasso Regression Prediction':y_pred_lasso,'Actual':y_test})
print(compare)
```

10- Use cross validation ( KFold cross_validation_score, default scoring method, n_splits=5)
to find the best performing regression algorithm (10 Marks).

```python
#Question 10:
from sklearn.model_selection import cross_val_score

cross_val = cross_val_score(reg,X_train,y_train, cv=5)
cross_val_ridge = cross_val_score(ridge,X_train,y_train, cv=5)
cross_val_lasso = cross_val_score(lasso,X_train,y_train, cv=5)

print(f'Linear Regression Cross Validation Score: {cross_val}')
print(f'Ridge Regression Cross Validation Score: {cross_val_ridge}')
print(f'Lasso Regression Cross Validation Score: {cross_val_lasso}')
```

11- Report on cv_result.mean and cv_result.std for the three models (10 Marks).


```python
#Question 11:

print(f'Linear Regression Cross Validation Score Mean: {cross_val.mean()}')
print(f'Linear Regression Cross Validation Score Standard Deviation: {cross_val.std()}')

print(f'Ridge Regression Cross Validation Score Mean: {cross_val_ridge.mean()}')
print(f'Ridge Regression Cross Validation Score Standard Deviation: {cross_val_ridge.std()}')

print(f'Lasso Regression Cross Validation Score Mean: {cross_val_lasso.mean()}')
print(f'Lasso Regression Cross Validation Score Standard Deviation: {cross_val_lasso.std()}')
```
